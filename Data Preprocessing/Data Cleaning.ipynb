```python
import pandas as pd
import numpy as np

```

```python
df=pd.read_csv('NFL.csv')
df.head()
```

```python
df = df.drop(labels=['Unnamed: 0'], axis=1)
df.head()
```

```python
missing_values_count=df.isnull().sum()
missing_values_count[0:10]
```

```python
total_cells = np.prod(df.shape)
print(total_cells)
missing_values_count_sum=missing_values_count.sum()
print(missing_values_count_sum)
prcent_missing_value=(missing_values_count_sum/total_cells)*100
print(prcent_missing_value)
```
